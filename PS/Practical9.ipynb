## Using sklearn API

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
melbourne_file_path = 'Melbourne_data/melb_data.csv'
melbourne_data = pd.read_csv(melbourne_file_path) 
melbourne_data.columns
```

```python
melbourne_data = melbourne_data.dropna(axis = 0)
```

```python
melbourne_data.shape
```

```python
melbourne_data.describe()
```

```python
melbourne_features = ['Rooms', 'Bathroom', 'Lattitude', 'Longtitude', 'YearBuilt']
```

```python
X = melbourne_data[melbourne_features]
X.describe()
```

```python
X.hist(bins = 50, figsize = (16,9))
plt.show()
```

```python
Y = melbourne_data['Price']
```

```python
Y.describe()
```

```python
Y = Y/1e6
```

```python
Y.describe()
```

```python
fig = plt.figure(figsize=(20,12))
plt.scatter(x = X['Lattitude'], y = X['Longtitude'],c = Y, alpha = 0.4)
cbar = plt.colorbar()
cbar.set_label('Price (1e6)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression(n_jobs = -1)
```

```python
model.fit(X_train, y_train)
```

```python
fig = plt.figure(figsize = (16,9))
plt.scatter(x = ['Rooms', 'Bathroom', 'Lattitude', 'Longtitude', 'YearBuilt'],y = model.coef_)
plt.grid(True)
```

```python
model.__dict__
```

```python
y_predict = model.predict(X_test)
```

```python
print(mean_absolute_error(y_test, y_predict))
```

```python
print(mean_squared_error(y_test, y_predict))
```

## Using statsmodel API

```python
import statsmodels.api as sm
```

```python
X = sm.add_constant(X)
```

```python
model = sm.OLS(Y, X).fit()
predictions = model.predict(X) 

print_model = model.summary()
print(print_model)
```
