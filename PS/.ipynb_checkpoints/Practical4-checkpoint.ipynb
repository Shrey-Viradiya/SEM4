```python
# importing Libraries

%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn

plt.style.use('ggplot')
```

```python
# using data from csv file

data = pd.read_csv('ages.csv')
ages = data['Age']
ids = data['Responder_id']

# computing mean, variance and standard deviation

mean = np.mean(ages)
variance = np.var(ages)
std = np.std(ages)
```

```python
# computing z score

z = np.array(list(map(lambda y : (y-mean)/std, ages)))
```

```python
# finding minimum and maximum

minimum = np.min(ages)
maximum = np.max(ages)
```

```python
# taking user input for number of bins

noOfbin= int(input("Enter the number of bins: "))
```

```python
# finding range and length of individual bin
rg = maximum - minimum
binLength =  rg/noOfbin
```

```python
# creating bins

bins = [minimum]
for i in range(noOfbin):
    bins.append(minimum+((i+1)*binLength))
```

```python
# creating histogram
# %matplotlib inline
fig=plt.figure(figsize=(16,9),dpi=300)
plt.hist(ages, bins=bins, edgecolor='black',log=False)

plt.title('Ages of Respondant')
plt.xlabel('Ages')
plt.ylabel('Total Respondants')

plt.tight_layout()
plt.show()
```

```python
# creating scatter plot

fig=plt.figure(figsize=(16,9),dpi=300)
plt.scatter(ages,z,cmap = "winter", edgecolor = 'red')
plt.title('Z values vs Ages')
plt.xlabel('Ages')
plt.ylabel('Z values')
plt.tight_layout()
plt.show()
```

```python
# density plot

fig=plt.figure(figsize=(16,9),dpi=300)
seaborn.distplot(ages,hist=False)
plt.show()
```
