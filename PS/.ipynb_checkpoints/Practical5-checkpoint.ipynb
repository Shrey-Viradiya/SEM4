```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
from scipy.stats import norm
from collections import Counter

plt.style.use('ggplot')
```

```python
data = pd.read_csv('StudentsPerformance.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
fig=plt.figure(figsize=(16,9),dpi=300)
seaborn.distplot(data['math score'],label='Mathematics',hist=False)
seaborn.distplot(data['reading score'],label='Reading',hist=False)
seaborn.distplot(data['writing score'],label='Writing',hist=False)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
print(data['math score'].std())
print(data['math score'].mean())
```

```python
fig=plt.figure(figsize=(16,9),dpi=300)
domain = np.linspace(0,100,1000)

seaborn.distplot(data['math score'],label='Mathematics Score Distribution',hist=True)
plt.plot(domain,norm.pdf(domain,data['math score'].mean(),data['math score'].std()),label='Mathematics Score Normal Distribution')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
fig=plt.figure(figsize=(16,9),dpi=300)
domain = np.linspace(0,100,1000)

seaborn.distplot(data['reading score'],label='Reading Score Distribution',hist=False)
plt.plot(domain,norm.pdf(domain,data['reading score'].mean(),data['reading score'].std()),label='Reading Score Normal Distribution')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
fig=plt.figure(figsize=(16,9),dpi=300)
domain = np.linspace(0,100,1000)

seaborn.distplot(data['writing score'],label='Writing Score Distribution',hist=False)
plt.plot(domain,norm.pdf(domain,data['writing score'].mean(),data['writing score'].std()),label='Writing Score Normal Distribution')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
def grade(x):
    if (x >= 90):
        return 'A'
    elif (x >= 75):
        return 'B'
    elif (x>=55):
        return 'C'
    elif (x>=35):
        return 'D'
    else:
        return 'F'
    
data['Math Grade'] = data['math score'].map(grade)
data['Writing Grade'] = data['reading score'].map(grade)
data['Reading Grade'] = data['writing score'].map(grade)
```

```python
grade_counter = Counter()

for g in data['Math Grade']:
    grade_counter.update(g)


plt.barh(list(grade_counter.keys()),list(grade_counter.values()))
plt.title('Mathematics Grades obtained by students')
plt.xlabel('No. of students')
plt.show()
```
