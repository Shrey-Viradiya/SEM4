## Break Statement

```python
for letter in 'Python':
    if letter == 'h':
      break
    print ('Current Letter :', letter)
print ('Breaked Letter :', letter)
```

## Continue statement

```python
for letter in 'Python':
    if letter == 'h':
      continue
    print ('Current Letter :', letter)
```

## Using other control structures in loops

```python
numbers = [11, 33, 55, 39, 55, 761, 37, 21, 23, 41, 13]
for num in numbers:
    if num%2==0:
        print('the list contains an even number')
        break
else:
    print('the list does not contain even number')
```
