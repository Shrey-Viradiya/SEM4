```python
#1) Assign function to a variable

def function(text): 
    return text.upper() + 'assignment'
  
print (function('variable ') )
  
function_to_variable = function 
  
print( function_to_variable('function ') )
```

```python
2.4#2) Define one function in another

def calc(a,b):
    print('a = {} b= {}' .format(a,b))
    
    def addition():
        print('{} + {} = {}'.format(a,b,(a+b)))
            
    addition()
    return

calc(10,20)

        
```

```python
#3) Pass one function as parameter in another function
def Function1(text):  
    return text.upper()  
  
def Function2(text):  
    return text.lower()  
  
def Demo(Another_function):  
    # storing the function in a variable  
    greeting = Another_function("Demonstration of function passed as an argument")  
    print(greeting) 
  
Demo(Function1)  
Demo(Function2) 
```

```python
#4) Possible that a function can return another function
def display():
    def message():
        return 'This is a function that returns another function'
    
    return message

#Call display() which inturn calls message()
#Returing function which is assigned to a variable
Function=display()

#Demonstrating the printing of messages 
print(Function())
print(Function)
```

# Scope and lifetime of variables


```python
#Scope and lifetime of variables

main_variable=100

def outer_function():
    main_variable=200
    
    
    def inner_function():
        main_variable=300
        print('Inner function {}'.format(main_variable))
        return
        
    inner_function()
    print('Outer function {}' .format(main_variable))
    return


outer_function()

print('Main function {}'.format(main_variable))
 
```

# Formal and Actual Parameters
### 1) Positional arguments

### 2) Keyword arguments

### 3) Default arguments

### 4) Variable length arguments


```python
# 1) Positional arguments
# 1 - to -1 mapping of arguments

# Function definition
def multiply(a,b):
    c=a*b
    return c


#Function call

x=10
y=20
d=multiply(x,y)
print('The Multiplication of {} X {} = {}'.format(x,y,d))
```

```python
#2) Keyword arguments

def item(item,price,x):
    price1 = x*price
    print('The price of {} {}  with per unit price of {}  is {}'.format(x,item,price,price1))

item('book',100,2)
item(price=200,x=20,item='pen')
```

```python
#3) Default arguments

def item(item,price,count=3):
    price1 = count*price
    print('The price of {} {}  with per unit price of {}  is {}'.format(count,item,price,price1))

item('book',100,2)
#Count has a default vakue
item(price=200,item='pen')
```

```python
#4) Variable length arguments
# *args for variable arguments
#*kwargs for passing keyworded, variable length dictionary argument list to the calling function.


# Format
# def function_name(positional_parameters, *args,*kwargs)

def addition(fargument, *args,**kwargs):
    print('Formal Argument',fargument)
    for i in args:
        print('i= ',i)

    s =sum(args)
    print('sum = ',s)
    for j in kwargs:
        print(kwargs[j])
        
    
    
print('With positional and *args parameters')
addition(100,20,30)


print('With positional and *args and **kwargs parameters')
addition('addition',10,20,30,sum='addition',multiply='multiply')
```

# Global vs. Local Variables
Variables that are defined inside a function body have a local scope, and those defined outside have a global scope.

This means that local variables can be accessed only inside the function in which they are declared, whereas global variables can be accessed throughout the program body by all functions.


```python
# This is a global variable.
total = 0 

# Function definition is here
def sum( arg1, arg2 ):
    # Add both the parameters and return them."
    total = arg1 + arg2; # Here total is local variable.
    print ("Inside the function local total : ", total)
    return

# Now you can call sum function
print(sum(10, 20))
print ("Outside the function global total : ", total )
```

```python
# This is a global variable.
total = 0 


# Function definition is here
def sum( arg1, arg2 ):
   # Add both the parameters and return them."
    global total
    total = arg1 + arg2; 
    print ("Inside the function local total : ", total)
    return

# Now you can call sum function
sum( 10, 20 )
print ("Outside the function global total : ", total )
```

## Multiple return values
### A python function can return multiple values 

```python
# This function returns a tuple
def function():
    str = "Python can return multiple values"
    x   = 100
    return str, x;  # Return tuple, we could also
                    # write (str, x)

#Call the function now
return_str, return_x = function() # Assign returned tuple
print(return_str)
print(return_x)
```
